1. Use matplotlib to plot the following equation:

y =x**2 âˆ’x+2

- You'll need to write the code that generates the x and y points.

- Add an anotation for the point 0, 0, the origin.

```python
x = np.linspace(-6, 6, 50)
y = x**2 - x + 2  

plt.plot(x, y)

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 8),
             arrowprops={'facecolor': 'blue'})

plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
n_rows= 2
n_cols= 2


x = np.linspace(0, 20, 50)
y = x**(0.5)
fig = plt.figure(figsize = (6, 4))
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y, c='black')
plt.title('âˆšx')

a = np.linspace(0, 20, 50)
b = a**3
fig = plt.figure(figsize = (6, 4))
plt.subplot(n_rows, n_cols, 2)
plt.plot(a, b, c='green')
plt.title('x^3')


d = np.linspace(0, 20, 50)
e = 2**d
fig = plt.figure(figsize = (6, 4))
plt.subplot(n_rows, n_cols, 3)
plt.plot(d, e, c= 'pink')
plt.title('2^x')


f = np.linspace(0, 20, 50)
g = 1/(f+1)
fig = plt.figure(figsize = (6, 5))
plt.subplot(n_rows, n_cols, 4)
plt.plot(f, g)
plt.title('1/(x+1)')

plt.subplots_adjust(left=None, 
                bottom=0.1, 
                right=None, 
                top=None, 
                wspace=None,
                hspace=None)

plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(10, 5))

plt.plot(x, y, marker = '*', c='black', label='$âˆšx$')
plt.plot(a, b, c='fuchsia', label='$x^3$')
plt.plot(d, e, c='green', label='$2^x$')
plt.plot(f, g, c='cyan', label='$1/(x+1)$')


plt.suptitle('Combined Plots', fontsize = 16)
plt.legend(loc='upper right')



plt.show()
```
