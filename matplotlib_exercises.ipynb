3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
n_rows= 2
n_cols= 2


x = np.linspace(0, 20, 50)
y1 = x**(0.5)
fig = plt.figure(figsize = (6, 4))
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y1, c='black')
plt.title('âˆšx')

x = np.linspace(0, 20, 50)
y2 = x**3
fig = plt.figure(figsize = (6, 4))
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2, c='green')
plt.title('x^3')


x = np.linspace(0, 20, 50)
y3 = 2**x
fig = plt.figure(figsize = (6, 4))
plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y3, c= 'pink')
plt.title('2^x')


x = np.linspace(0, 20, 50)
y4 = 1/(x+1)
fig = plt.figure(figsize = (6, 5))
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y4)
plt.title('1/(x+1)')

plt.subplots_adjust(left=None, 
                bottom=0.1, 
                right=None, 
                top=None, 
                wspace=None,
                hspace=None)

plt.tight_layout()

plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(6, 4))
x = np.linspace(0, 20, 50)

y1 = x**(0.5)
plt.plot(x, y1, marker = 'h', c='black', label='$âˆšx$')


y2 = x**3
plt.plot(x, y2, c='green')
plt.plot(x, y2, marker = '.', c='fuchsia', label='$x^3$')

y3 = 2**x
plt.plot(x, y3, marker = 'v',  c= 'pink')
plt.plot(x, y3, c='green', label='$2^x$')



y4 = 1/(x+1)
plt.plot(x, y4)
plt.plot(x, y4, marker = '*', c='cyan', label='$1/(x+1)$')


plt.xlim(0,20)
plt.ylim(0,15)
plt.title('Combined Plots', fontsize = 16)
plt.legend(loc='upper right')



plt.show()
```
