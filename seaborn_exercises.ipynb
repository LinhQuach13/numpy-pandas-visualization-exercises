# 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
#chiptole database
orders_df = pd.read_sql('select * from orders', url)
```

```python
orders_df
```

```python
#make item_price a float
orders_df['item_price'] = orders_df['item_price'].str.replace('$','').str.replace(',','').astype('float')
```

```python
orders_df
```

```python
item_quantity= orders_df.groupby('item_name').item_price.sum().nlargest(4)
```

```python
item_quantity
```

```python
top_4= item_quantity.reset_index()
```

```python
sns.barplot(x='item_name', y= 'item_price', data=top_4)
```
