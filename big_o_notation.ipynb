```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import math
```

```python
plt.figure(figsize = (10, 8))

plt.title('Big O Notation', fontsize = 16)
plt.xlabel('Elements')
plt.ylabel('Operations')

n = np.array([1,2,3,4,5])
y = 0*n+1
plt.plot(n, y, c='black', label='$O(1)$')

n = np.array([1,2,3,4,5])
y2 = [math.log(x) for x in n]
plt.plot(n, y2, c='darkmagenta', label='$O(log n)$')


n = np.array([1,2,3,4,5])
y3 = n
plt.plot(n, y3, c='cyan', label='$O(n)$')


n = np.array([1,2,3,4,5])
y4 = [n* math.log(x) for x in n]
plt.plot(n, y4, c='red', label='$O(n log n)$')


n = np.array([1,2,3,4,5])
y5 = n**2
plt.plot(n, y5, c='forestgreen', label='$O(n^2)$')


n = np.array([1,2,3,4,5])
y6 =2*n
plt.plot(n, y6, c='sandybrown', ls=':', label='$O(2^n)$')

n= np.array([1,2,3,4,5])
y7 = [math.factorial(x) for x in n]
plt.plot(n, y7, c='deeppink', label='$O(n!)$')


n = np.array([1,2,3,4,5])
y8 = n**n 
plt.plot(n, y8, c='darkslategray', label='$O(n^n)$')

plt.legend()

plt.show()
```
